```python
import train as tn
reload(tn)

import plotting
reload(plotting)

import util as ut

import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

import numpy as np
```

```python
path,name = "./classifiers", "class_tnevts_iso_out"
fitter = tn.IO.load(name,path,nodata=True)
```

```python
def featureImpPlot (features, features_importance, title) :
    y_position = np.arange(len(features))

    plt.bar(y_position, features_importance*100, align='center', alpha=0.5)
    plt.xticks(y_position, features,rotation='90')
    plt.ylabel('feature importance [%]')
    plt.title("trained class: " + title)
    plt.show()
```

```python
classes = ['class','recoPt','recoNjets2p5']
"""
Gini importance is used by sklearn
"""
for c in classes :
    feature, feature_imp = fitter.featureImportance(c)
    print(feature_imp)
    featureImpPlot(feature,feature_imp,title=c)
```
